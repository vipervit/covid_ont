```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
import sys
import os
sys.path.append(os.getenv('DEV_HOME'))
sys.path.append(os.getenv('SITES_HOME'))
import covid_ont
from covid_ont import dataset_read, dataset_get
```

```python
# df_cases=dataset_get('Cases by PHU')
```

```python
df_cases=dataset_read('Cases by PHU')
```

```python
df_cases.shape
```

```python
df_cases.tail()
```

```python
df_cases['Date']=pd.to_datetime(df_cases['Date'])
df_cases.head()
```

```python
df_cases.columns=df_cases.columns.str.replace('_', ' ')
```

```python
top_5=list(df_cases.mean().sort_values(ascending=False).head(6).index.drop('Total'))
```

```python
top_5
```

```python
df_cases_5=df_cases.drop([col for col in df_cases.columns if col not in top_5 and col!='Date'], axis=1)
df_cases_5.set_index('Date', inplace=True)
df_cases_5
```

```python
fig, ax = plt.subplots(figsize=(15, 6))
ax.spines['left'].set_visible(False)
ax.spines['bottom'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
df_cases_5.plot(ax=ax).set_title('Daily New Cases in 5 Most Infected Public Health Units', fontsize=20)
```
