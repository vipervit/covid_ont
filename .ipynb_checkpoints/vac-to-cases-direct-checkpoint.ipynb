```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df_vac_src=pd.read_csv('https://data.ontario.ca/dataset/752ce2b7-c15a-4965-a3dc-397bf405e7cc/resource/8a89caa9-511c-4568-af89-7f2174b4378c/download/vaccine_doses.csv')
```

```python
df_vac_src.head()
```

```python
df_cases_src=pd.read_csv('https://data.ontario.ca/dataset/f4f86e54-872d-43f8-8a86-3892fd3cb5e6/resource/8a88fe6d-d8fb-41a3-9d04-f0550a44999f/download/daily_change_in_cases_by_phu.csv')
```

```python
df_cases_src.head()
```

```python
df_cases=df_cases_src[['Date', 'Total']]
df_cases.columns=['Date', 'New Cases']
df_cases.head()                  
```

```python
df_vac=df_vac_src[['report_date', 'total_individuals_fully_vaccinated']]
df_vac.columns=['Date', 'Fully Vaccinated Total']
df_vac
```

```python
df_fully=df_vac[['Date', 'Fully Vaccinated Total']].dropna().reset_index(drop=True)
df_fully.drop(df_fully.loc[:23].index, inplace=True)
df_fully.reset_index(drop=True, inplace=True)
df_fully['Fully Vaccinated Total']=df_fully['Fully Vaccinated Total'].str.replace(',', '')
df_fully['Fully Vaccinated Total']=df_fully['Fully Vaccinated Total'].astype(int)
df_fully.fillna(0, inplace=True)
df_fully.head()
```

```python
df_plot=pd.merge(df_vac, df_cases, on='Date')
df_plot['Fully Vaccinated Total']=df_plot['Fully Vaccinated Total'].str.replace(',', '')
df_plot.fillna(0, inplace=True)
df_plot['Fully Vaccinated Total']=df_plot['Fully Vaccinated Total'].astype(int)
# df_plot['Fully Vaccinated Total']=df_plot['Fully Vaccinated Total'].apply(lambda x: x / 100)
df_plot.tail()
```

```python
df_plot.head()
```

```python
fig, ax1 = plt.subplots(1, 1, figsize=(20,5))
df_plot.plot(ax=ax1, x='Date', y='New Cases', color='red', label='New Cases', xlabel='')
ax2=ax1.twinx()
df_plot.plot(ax=ax2, x='Date', y='Fully Vaccinated Total', color='green', label='Fully Vaccinated')
ax1.set_ylabel('New Cases')
ax2.set_ylabel('Vaccinations')
ax1.legend()
ax2.legend()
```
